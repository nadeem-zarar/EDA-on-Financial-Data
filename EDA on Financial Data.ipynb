```python
import yfinance as yf
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# define the yahoo finance ticker for nifty50
nifty_ticker = '^NSEI'
start_date = dt.date(2019,1,1)
end_date = dt.date(2024,5,30)
```

```python
nifty_levels_data = yf.download(nifty_ticker, start=start_date, end=end_date)
```

```python
nifty_levels_data.head()
```

```python
nifty_levels_data.tail()
```

```python
nifty_levels_data.info()
```

```python
nifty_levels_data.columns
```

```python
plt.figure(figsize = (12,8))

sns.lineplot(x = nifty_levels_data.index, y = nifty_levels_data[('Close', '^NSEI')], color = 'purple')
plt.title('Nifty 50 Closing Price')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
```

```python
stock_ticker_list = (nifty50['Symbol'] + '.NS').tolist()
stock_ticker_list
```

```python
stock_data = yf.download(stock_ticker_list, start=start_date, end=end_date)
```

```python
close_stock_data = stock_data['Close'] # only selected close metric
```

```python
close_stock_data.head()
```

## Visualizing all stocks Close

```python
close_stock_data.columns
```

```python
plt.figure(figsize = (14,7))
for ticker in close_stock_data.columns:
    sns.lineplot(x=close_stock_data.index, y=close_stock_data[ticker], label=ticker)
```

```python
plt.figure(figsize = (14,7))
banks = ['AXISBANK.NS', 'HDFCBANK.NS', 'ICICIBANK.NS', 'INDUSINDBK.NS', 'KOTAKBANK.NS']
for ticker in banks:
    sns.lineplot(x=close_stock_data.index, y=close_stock_data[ticker], label=ticker)
```

```python
plt.figure(figsize = (14,7))
banks = ['TCS.NS', 'WIPRO.NS']
for ticker in banks:
    sns.lineplot(x=close_stock_data.index, y=close_stock_data[ticker], label = ticker)
```

```python
open_stock_data = stock_data['Open'] # only selected for open metric
open_stock_data
```

```python
plt.figure(figsize = (14,7))
banks = ['TCS.NS', 'WIPRO.NS']
for ticker in banks:
    sns.lineplot(x=open_stock_data.index, y=open_stock_data[ticker], label = ticker)
```

## Creating Visual for Specific Stock

#### xs method

```python
stock_data.xs(key = 'TCS.NS', axis = 1, level = 1, drop_level = False)
```

```python
tcs_stock = stock_data.xs(key = 'TCS.NS', axis = 1, level = 1)
tcs_stock
```

```python
plt.figure(figsize = (14,10))

sns.lineplot(data = tcs_stock[['High', 'Low']])

plt.title('Stock Metrics for TCS.NS')
plt.xlabel('Date')
plt.ylabel('Price')

plt.show()
```
